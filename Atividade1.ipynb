Questão 9

```python
for col in data.columns:
    if is_object_dtype(data[col]):
        data[col].fillna(data[col].mode()[0],inplace=True)
    elif is_numeric_dtype(data[col]):
        data[col].fillna(data[col].mean(),inplace=True)
```

```python
data.isna().sum()
```

```python
for column in data.columns:
    if is_object_dtype(data[column]) and data[column].value_counts().shape[0] >= 3:
        data.drop(columns=column, inplace=True)
```

```python
data.head()
```

Questão 11

```python
lista_categorica=[]

for column in data.columns:
    if is_object_dtype(data[column]):
        lista_categorica.append(column)
```

```python
# fit and transform data
data_encoded = pd.get_dummies(data, columns=lista_categorica)
data_encoded
```

Questão 12

```python
data2 = data_encoded

data2.boxplot(figsize=(20,3))
```

```python
from sklearn.neighbors import LocalOutlierFactor
from numpy import where

lof = LocalOutlierFactor(n_neighbors=3, p=2)

y_pred = lof.fit_predict(data2)
y_pred
```

```python
lofs_index = where(y_pred == -1)
lofs_index
```

```python
# Vizualizando 10 primeiros outliers

values = data2.iloc[lofs_index]
values.head(10)
```

```python
lofs_index[0]
```

```python
# Número de Outliers
print(f'Existem {len(lofs_index[0])} outliers')
values.index
```

```python
# Removendo os outliers

print(f'Numero de elementos antes da remoção: {data2.shape[0]}')
data2 = data2.drop(values.index)
print(f'Numero de elementos depois da remoção: {data2.shape[0]}')
```

Questão 14

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Padronizando antes de aplicar PCA

scaler = StandardScaler()
data2_padronizado = scaler.fit_transform(data2)
```

```python
# Aplicando PCA

pca = PCA(n_components=2)

projected = pca.fit_transform(data2_padronizado)

component_names = ['component {}'.format(i) for i in range(len(pca.components_))]

components_data2 = pd.DataFrame(data=pca.components_, index=component_names, columns=data2.columns)
components_data2.head()
```
